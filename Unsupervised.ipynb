```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans,AgglomerativeClustering, DBSCAN
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
X,_=make_blobs(n_samples=300,centers=4,random_state=42)
inertia_values=[]
K_values=range(1,11)
for k in K_values:
    model =KMeans(n_clusters= k,n_init=10,random_state=42)
    model.fit(X)
    inertia_values.append(model.inertia_)
plt.plot(K_values,inertia_values,marker='o')
plt.grid()
plt.show()
```

```python
X, y = make_blobs(n_samples=300, centers=4, random_state=42)
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
plt.figure(figsize=(6, 4))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels, palette='viridis', s=50, alpha=0.8, edgecolor='k')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200)
plt.title('K-Means Clustering')
plt.show()
```

```python
X,_=make_blobs(n_samples=300,centers=4,random_state=42)
hierarchical =AgglomerativeClustering(n_clusters= 4,linkage='ward')
hierarchical_predict=hierarchical.fit_predict(X)
hierarchical_predict
```

```python
X,y=make_blobs(n_samples=300,centers=4,random_state=42)
hierarchical =AgglomerativeClustering(n_clusters= 4,linkage='ward')
hierarchical_predict=hierarchical.fit_predict(X)
plt.figure(figsize=(6, 4))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=hierarchical_predict, palette='viridis', s=50, alpha=0.8, edgecolor='k')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
Z = linkage(X, method='ward')  
dendrogram(Z)
plt.show()
```

```python
X,y=make_blobs(n_samples=300,centers=4,random_state=42)
model =DBSCAN(eps=1.5, min_samples=4)
model=model.fit_predict(X)
plt.figure(figsize=(6, 4))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=model, palette='viridis', s=50, alpha=0.8, edgecolor='k')
plt.show()
```
